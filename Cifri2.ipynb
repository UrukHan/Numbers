```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import pytesseract
from google.colab import files
%matplotlib inline
```

```
def showImg(name):
  plt.imshow(name, cmap = 'gray', interpolation = 'bicubic')
  plt.show()
```

```
cifr = cv2.imread('/content/nnn.jpg')
showImg(cifr)
```

```
cifri = cv2.cvtColor(cifr, cv2.COLOR_BGR2GRAY)
ret, cifri = cv2.threshold(cifri, 160, 255, cv2.THRESH_BINARY)
showImg(cifri)
```

```
text = pytesseract.image_to_string(cifri, lang = 'ara_number', config = '--psm 1')
print(text)
```
